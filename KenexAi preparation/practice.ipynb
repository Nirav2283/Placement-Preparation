```python
def vow_replace(string , vowel):
    vowels = "aeiou"
    result = ""
    for char in string:
        if char in vowels:
            result += vowel
        else:
            result += char
    return result

ans = vow_replace("nirav" , "e")

print(ans)
```

```python
def mapping(letters):
    result = {}
    for l in letters:
        result[l] = l.upper()
    return result

mapping(["n" , "s"])
```

```python
def equality(a,b,c):
    if a == b == c:
        return 3
    elif a == b or b == c or c == a:
        return 2
    else:
        return 0


equality(2,1,3)
```

```python
def pronic(start ,end , num):
    for i in range(start , end + 1):
        if i*(i+1) == num:
            return i;
    return "no number found"


pronic(1,100 , 2)

```
